# Matplotlib绘图

## 绘制函数图形

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成数据
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
```

```python
# 绘图
plt.plot(x, y)
plt.show()
```

## 绘制复杂图形

```python
# 导入库函数　　　
#import numpy as np　　　
#import matplotlib.pyplot as plt
```

```python
# 生成数据　　　
x = np.arange(0, 4 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
# 绘图
plt.plot(x, y_sin)
plt.plot(x, y_cos, "--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin and cos")
plt.legend (["sin", "cos"])
plt.show()
```

## 绘制散点图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成数据
x = np.random.randn(200)
y = np.random.randn(200)
# 绘制散点图
plt.scatter(x, y)
plt.show()
```

## 不同颜色区分

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.randn(200)
y1 = np.random.randn(200)
x2 = np.random.randn(200) + 3
y2 = np.random.randn(200) + 3
```

```python
# 绘制散点图
plt.scatter(x1, y1, marker='o')
plt.scatter(x2, y2, marker='s')
plt.show()
```
